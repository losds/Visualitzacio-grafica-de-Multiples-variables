###  Exercici 2

Repeteix l'exercici 1 amb el dataset que disposem en el repositori de GitHub PRE-PROCESSING-DATA, movies.dat 

```python
heder = ['id', 'titulo', 'genero']
movies = pd.read_table('movies.dat', sep = '::',header = None,names = heder)
movies.head()
```

```python
movies['a√±o']=movies["titulo"].str[-5:-1]
```

```python
movies.describe(include='all')
```

```python
movies.head()
```

```python
movies.head()
```

```python
generos=[]
```

```python
for x in movies.genero:
    generos.extend(x.split('|'))
```

```python
generos
```

```python
generos = pd.unique(generos)
```

```python
generos
```

```python
datos = pd.DataFrame(np.zeros((len(movies), len(generos))), columns=generos)
```

```python
datos
```

```python
movies
```

Relleno la matriz de generos

```python
for posicion, genero in enumerate(movies.genero):
    index = datos.columns.get_indexer(genero.split('|'))
    datos.iloc[posicion, index] = 1
```

```python
datos
```

```python
datos
```

```python
pd.set_option('display.max_columns', None)
```

```python
datos
```

```python
moviestab= movies.join(datos)
```

```python
moviestab
```

```python
moviestab.iloc[:,4:].sum().sort_values(ascending=False).plot.bar()
```

Visualizo las peliculas por el genero, la mayoria de las peliculas son de  drama luego va comedia y las peliculas de fantasia son los menos que aparecen son Film-Noir

```python
pivot=moviestab.iloc[:,3:].groupby('a√±o').sum()
pivot
```

```python
pivot["a√±o"]=pivot.index
```

```python
unpivot=pd.melt(pivot, id_vars=['a√±o'],var_name="genero", value_name="cantidad")
unpivot
```

```python
g=sns.relplot(x="a√±o", y="cantidad", hue="genero", data=unpivot)
g.fig.set_figwidth(50)
g.fig.set_figheight(15)
```

Como se puede observar en la grafica hasta el a√±o 1986 la cantidad por genero no superaba 25, a finales de 89 hay un crecimiento muy grande en la cantidad de peliculas. Tambien se puede observar que cada a√±o dramas  lideran top 5

```python
unpivot.groupby("a√±o").sum().plot.bar(figsize=(15, 20))
```

Se puede ver como evoluciona la cantidad total de peliculas a√±os 98,99 tienen el pico

```python
plt.figure(figsize=(30,20))
g=sns.lineplot(data=pivot)
```

En el grafico con todos los generos es deficil sacar la info se solapan voy a sacar unos cuantos generos y comprarar como se queda
